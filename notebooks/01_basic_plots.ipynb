```python
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use("ggplot")
```

# Generate basic plots

This notebook creates some of the basic plots used in the [documentation](https://nivanorge.github.io/lake_liming_app/).

## 1. Curves for instantaneous dissolution

```python
prod = "Standard Kalk Kat3"
```

```python
data_path = r"../data/lime_products.xlsx"
df = pd.read_excel(data_path, sheet_name="Sheet1")
```

```python
id_list = df[prod].tolist()[4:9]

plt.plot([4, 4.5, 5, 5.5, 6], id_list, "ro-")
plt.title(f"Momentanoppløsningskurve for\n'{prod}'", fontsize=14)
plt.ylabel("Momentanoppløsning (%)")
plt.xlabel("pH (-)")
plt.tight_layout()

plt.savefig(
    r"../docs/images/instantaneous_dissolution_example.png",
    dpi=200,
    bbox_inches="tight",
)
```

## 2. Curves for overdosing

```python
od_list = df[prod].tolist()[9:]

plt.plot([10, 20, 35, 50, 85], od_list, "ro-")
plt.title(f"Overdoseringskurve for\n'{prod}'", fontsize=14)
plt.ylabel("Overdoseringsfaktor (-)")
plt.xlabel("Dose (mg/l)")
plt.tight_layout()

plt.savefig(r"../docs/images/overdosing_example.png", dpi=200, bbox_inches="tight")
```

## 3. Flow profiles

```python
q_df = pd.read_excel(r"../data/flow_typologies.xlsx", index_col=0)
q_df.columns = [col.capitalize() for col in q_df]
q_df.plot()
plt.xlabel("Måned")
plt.ylabel("Faktor (-)")
plt.tight_layout()
plt.savefig(r"../docs/images/flow_typologies.png", dpi=200, bbox_inches="tight")
```
